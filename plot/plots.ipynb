```python
def plot_data(points):
    plt.figure(figsize=(8, 5), layout="constrained")
    for key in points.keys():
        data = np.array(points[key])
        x = range(1, len(data) + 1)
        y = data[:, 2]
        plt.plot(x, y, label=key)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title("")
    plt.legend(loc="lower right", ncols=2, borderaxespad=0.1)

plot_data(train_acc)
```

```python
def plot_bar_auc(train, val, names):
    n_groups = len(train)

    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.35
    opacity = 0.8

    rects1 = plt.bar(index, train, bar_width, alpha=opacity, label="Train")

    rects2 = plt.bar(index + bar_width, val, bar_width, alpha=opacity, label="Val")

    plt.xlabel("Models")
    plt.ylabel("AUC")
    plt.title("")
    plt.xticks(index + bar_width / 2, names, rotation=-45, ha="left", rotation_mode="anchor")
    plt.legend(loc="lower right", ncols=1, borderaxespad=0.1)

    # plt.tight_layout()
    plt.grid(axis="y", linestyle = '--')
    plt.show()

def plot_bar_runtime(runtime, names):
    fig = plt.figure()

    plt.bar(names,runtime, color="#3DDC84")

    plt.xlabel("Models")
    plt.ylabel("Runtime (min.)")
    plt.title("")
    plt.xticks(names, rotation=-45, ha="left", rotation_mode="anchor")

    # plt.tight_layout()
    plt.grid(axis="y", linestyle = '--')
    plt.show()
```

```python
# data to plot

train = df["train/auc (last)"]
val = df["val/auc (last)"]
names = df[["Tags", "params/model"]].apply(": ".join, axis=1)

plot_bar_auc(train, val, names)
```
